```python
import requests
import pandas as pd
def all_players(season_list,s_type):
    seasons = []

    for seas in season_list:
        season =  int(seas) -1
        trail = str((season+1)%100)
        if len(trail) == 1:
            trail = '0'+trail
        season_s = str(season)+'-'+trail
        seasons.append(season_s)
    #print(seasons)
    frames = []
    for season in seasons:
        wowy_url = "https://api.pbpstats.com/get-totals/nba"
        wowy_params = {

                "Season": season,
                "SeasonType": s_type,
                "Type": "Player" # Player stats
            }
        wowy_response = requests.get(wowy_url, params=wowy_params)
        wowy = wowy_response.json()
        player_stats= wowy["multi_row_table_data"]
        players = pd.DataFrame(player_stats)
        players['year'] = int(season[0:4])+1
        frames.append(players)
        #print(players.head(1))
        print(season+ ' done')
    
    #print(len(player_stats))
    return pd.concat(frames)
rs = all_players([i for i in range(2001,2024)],s_type='Regular Season')
```

```python
rs.to_csv('pbp_rs_totals.csv',index = False)
```
